```python
#id,genres,title,overview,companies from movies.metadata
# cast, crew from credits
#keywords
credits = pd.read_csv("/kaggle/input/the-movies-dataset/credits.csv")
keywords = pd.read_csv("/kaggle/input/the-movies-dataset/keywords.csv")
movies_metadata = pd.read_csv("/kaggle/input/the-movies-dataset/movies_metadata.csv")
```

```python
movies_metadata = movies_metadata[['id','genres','original_title','overview','production_companies']]
credits = credits[['id','cast','crew']]
keywords = keywords[['id','keywords']]
```

```python
movies_metadata['id']=movies_metadata['id'].strip().astype('Int64')
keywords['id']=keywords['id'].astype(int)

merged_data = pd.concat((keywords,movies_metadata),axis = 1)
merged_data = pd.concat((merged_data,credits),axis = 1)

merged_data.head()
```
